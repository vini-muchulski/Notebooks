```python

import pandas as pd
import numpy as np
import yfinance as yf
```

```python

ativos = ["BPAC11.SA","ITUB3.SA","VALE3.SA","EGIE3.SA","WEGE3.SA","^BVSP"]

data_inicio = pd.to_datetime("2019-01-01")
data_fim = pd.to_datetime("2023-12-31")

carteira = yf.download(ativos,start=data_inicio,end=data_fim)["Adj Close"]
carteira
```

# Calculando retornos

```python
retornos = carteira.pct_change()
retornos.dropna(inplace=True)
retornos
```

```python
retornos.var(axis=0)
```

# Correlaçao

```python
retornos.corr("pearson")
```

# Covariancia


```python
retornos.cov()
```
